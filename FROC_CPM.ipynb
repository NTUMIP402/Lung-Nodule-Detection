```python
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
res_fps = np.load('CPM_Results_sCaps_80/sCaps_80_all_fps.npy')
res_sens = np.load('CPM_Results_sCaps_80/sCaps_80_all_sens.npy')


plt.rcParams["font.family"] = "Times New Roman"
plt.plot(res_fps, res_sens, color='red', label='sCaps, CPM = 0.864', linestyle=':')

plt.legend(loc='lower right', fontsize=16)
plt.xlabel('FPs per scan')
plt.ylabel('Sensitivity')
plt.title('FROC Curve')
plt.savefig('FROC_sCaps.png', dpi=300)
```
