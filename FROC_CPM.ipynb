```python
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline

res_fps = np.load('OSAF_YOLOv3_fps.npy')
res_sens = np.load('OSAF_YOLOv3_sens.npy')


plt.rcParams["font.family"] = "Times New Roman"
plt.plot(res_fps, res_sens, color='red', label='OSAF_YOLOv3, CPM = 0.905', linestyle=':')

plt.legend(loc='lower right', fontsize=16)
plt.xlabel('FPs per scan')
plt.ylabel('Sensitivity')
plt.title('FROC Curve')
plt.savefig('FROC_sCaps.png', dpi=300)
```
